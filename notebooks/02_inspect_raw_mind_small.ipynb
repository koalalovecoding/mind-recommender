## Step 2. Inspect raw MIND-small files

This section verifies that the required raw MIND-small files are available in the expected project directories. We first check the existence and file size of `behaviors.tsv` and `news.tsv` for both the training and development splits.

```python
from pathlib import Path

import pandas as pd

# Define the root directory for the raw MIND-small dataset.
# The notebook is inside the notebooks/ folder, so "../data/raw" points back to the project-level data/raw directory.
DATA_DIR = Path("../data/raw")

train_behaviors_path = DATA_DIR / "MINDsmall_train" / "behaviors.tsv"
train_news_path = DATA_DIR / "MINDsmall_train" / "news.tsv"

dev_behaviors_path = DATA_DIR / "MINDsmall_dev" / "behaviors.tsv"
dev_news_path = DATA_DIR / "MINDsmall_dev" / "news.tsv"

# Check whether each required raw file exists and print its file size in bytes.
# This helps verify that the dataset was downloaded and placed correctly.
# It also helps detect failed downloads, such as tiny XML error files instead of real TSV files.
for path in [
    train_behaviors_path,
    train_news_path,
    dev_behaviors_path,
    dev_news_path,
]:
    print(path, "exists:", path.exists(), "size:", path.stat().st_size if path.exists() else None)
```

```python
# Define column names for the MIND behaviors file.
# The raw behaviors.tsv file does not include a header row, so we manually assign column names.
behavior_cols = [
    "impression_id",  # Unique identifier for each impression event.
    "user_id",        # Anonymous user identifier.
    "time",           # Timestamp of the impression event.
    "history",        # Previously clicked news articles for this user.
    "impressions",    # Candidate news articles shown in this impression, with click labels.
]

# Define column names for the MIND news file.
# The raw news.tsv file also does not include a header row.
news_cols = [
    "news_id",            # Unique identifier for each news article.
    "category",           # High-level news category.
    "subcategory",        # More specific news subcategory.
    "title",              # News title.
    "abstract",           # News abstract or summary.
    "url",                # Original news URL.
    "title_entities",     # Entities extracted from the title.
    "abstract_entities",  # Entities extracted from the abstract.
]

# Load the training behavior logs.
# sep="\t" means the file is tab-separated.
# names=behavior_cols assigns the manually defined column names.
# header=None tells pandas that the file does not have an existing header row.
train_behaviors = pd.read_csv(
    train_behaviors_path,
    sep="\t",
    names=behavior_cols,
    header=None,
)

# Load the development/validation behavior logs.
dev_behaviors = pd.read_csv(
    dev_behaviors_path,
    sep="\t",
    names=behavior_cols,
    header=None,
)

# Load the training news metadata.
train_news = pd.read_csv(
    train_news_path,
    sep="\t",
    names=news_cols,
    header=None,
)

# Load the development/validation news metadata.
dev_news = pd.read_csv(
    dev_news_path,
    sep="\t",
    names=news_cols,
    header=None,
)

# Print the shape of each DataFrame.
# The shape is shown as (number of rows, number of columns).
# This confirms that the files were loaded correctly and have the expected structure.
print("train_behaviors:", train_behaviors.shape)
print("dev_behaviors:", dev_behaviors.shape)
print("train_news:", train_news.shape)
print("dev_news:", dev_news.shape)
```

### Missing-value check

We check missing values in the behavior logs and news metadata. Missing `history` values may indicate users without prior clicks, while missing `abstract` values may indicate news articles without summaries. These cases should be documented before building interaction tables or using text features.

```python
# Missing-value check for all loaded raw tables.
# This is important before parsing and modeling because some fields may be empty.
# For example:
# - history can be missing if a user has no previous clicked news.
# - abstract can be missing if a news article does not provide a summary.
# These missing values are not necessarily errors, but we need to know they exist.

print("Train behaviors missing values:")
display(train_behaviors.isna().sum())

print("Dev behaviors missing values:")
display(dev_behaviors.isna().sum())

print("Train news missing values:")
display(train_news.isna().sum())

print("Dev news missing values:")
display(dev_news.isna().sum())
```

```python
# Print example impression strings together with the corresponding user_id.
# Each impression string contains the news articles shown to the user in one impression event,
# where "-1" means clicked and "-0" means exposed but not clicked.

print("Example impression strings with user_id:")

for _, row in train_behaviors[["user_id",  "impressions"]].dropna().head(5).iterrows():
    print("user_id:", row["user_id"])
    print("impressions:", row["impressions"])
    print("-" * 80)


# Print example history strings together with the corresponding user_id.
# The history string contains news articles clicked by the user before the current impression event.

print("\nExample history strings with user_id:")

for _, row in train_behaviors[["user_id", "history"]].dropna().head(5).iterrows():
    print("user_id:", row["user_id"])
    print("history:", row["history"])
    print("-" * 80)
```

```python
# Count how many candidate news articles were shown in each impression event.
# The impressions column is a space-separated string such as:
# "N55689-1 N35729-0"
# Each token represents one exposed news article with a click label.

train_behaviors["num_impressions"] = train_behaviors["impressions"].fillna("").apply(
    lambda x: len(x.split()) if x else 0
)

dev_behaviors["num_impressions"] = dev_behaviors["impressions"].fillna("").apply(
    lambda x: len(x.split()) if x else 0
)

# Summarize the distribution of the number of candidate news articles per impression.
print("Train impression candidate count:")
display(train_behaviors["num_impressions"].describe())

print("Dev impression candidate count:")
display(dev_behaviors["num_impressions"].describe())
```

```python
# Count clicked and non-clicked items in one impression string.
# Example:
# "N55689-1 N35729-0 N12345-0"
# means:
# - N55689 was exposed and clicked
# - N35729 and N12345 were exposed but not clicked
def count_clicks(impression_string):
    # If the impression string is missing, return zero clicks and zero non-clicks.
    if pd.isna(impression_string):
        return 0, 0

    clicked = 0
    non_clicked = 0

    # Each token has the format "news_id-click_label",
    # where "-1" means clicked and "-0" means not clicked.
    for token in impression_string.split():
        if token.endswith("-1"):
            clicked += 1
        elif token.endswith("-0"):
            non_clicked += 1

    # Return a pair: (number of clicked items, number of non-clicked exposed items).
    return clicked, non_clicked


# Apply the counting function to every impression string in the training and development sets.
train_click_counts = train_behaviors["impressions"].apply(count_clicks)
dev_click_counts = dev_behaviors["impressions"].apply(count_clicks)

# Extract the first element of each pair as the number of clicked items.
train_behaviors["num_clicked"] = train_click_counts.apply(lambda x: x[0])

# Extract the second element of each pair as the number of exposed but non-clicked items.
train_behaviors["num_non_clicked"] = train_click_counts.apply(lambda x: x[1])

# Do the same extraction for the development set.
dev_behaviors["num_clicked"] = dev_click_counts.apply(lambda x: x[0])
dev_behaviors["num_non_clicked"] = dev_click_counts.apply(lambda x: x[1])

# Print total clicked and non-clicked exposed items across all impression events.
print("Train clicked items:", train_behaviors["num_clicked"].sum())
print("Train non-clicked exposed items:", train_behaviors["num_non_clicked"].sum())

print("Dev clicked items:", dev_behaviors["num_clicked"].sum())
print("Dev non-clicked exposed items:", dev_behaviors["num_non_clicked"].sum())
```
