```python
from pathlib import Path

import pandas as pd

DATA_DIR = Path("../data/raw")

TRAIN_DIR = DATA_DIR / "MINDsmall_train"
DEV_DIR = DATA_DIR / "MINDsmall_dev"

train_behaviors_path = TRAIN_DIR / "behaviors.tsv"
train_news_path = TRAIN_DIR / "news.tsv"

dev_behaviors_path = DEV_DIR / "behaviors.tsv"
dev_news_path = DEV_DIR / "news.tsv"

for path in [
    train_behaviors_path,
    train_news_path,
    dev_behaviors_path,
    dev_news_path,
]:
    print(path, "exists:", path.exists(), "size:", path.stat().st_size if path.exists() else None)
```

```python
behavior_cols = [
    "impression_id",
    "user_id",
    "time",
    "history",
    "impressions",
]

news_cols = [
    "news_id",
    "category",
    "subcategory",
    "title",
    "abstract",
    "url",
    "title_entities",
    "abstract_entities",
]

train_behaviors = pd.read_csv(
    train_behaviors_path,
    sep="\t",
    names=behavior_cols,
    header=None,
)

dev_behaviors = pd.read_csv(
    dev_behaviors_path,
    sep="\t",
    names=behavior_cols,
    header=None,
)

train_news = pd.read_csv(
    train_news_path,
    sep="\t",
    names=news_cols,
    header=None,
)

dev_news = pd.read_csv(
    dev_news_path,
    sep="\t",
    names=news_cols,
    header=None,
)

print("train_behaviors:", train_behaviors.shape)
print("dev_behaviors:", dev_behaviors.shape)
print("train_news:", train_news.shape)
print("dev_news:", dev_news.shape)
```

```python
train_behaviors["num_impressions"] = train_behaviors["impressions"].fillna("").apply(
    lambda x: len(x.split()) if x else 0
)

dev_behaviors["num_impressions"] = dev_behaviors["impressions"].fillna("").apply(
    lambda x: len(x.split()) if x else 0
)

print("Train impression candidate count:")
display(train_behaviors["num_impressions"].describe())

print("Dev impression candidate count:")
display(dev_behaviors["num_impressions"].describe())
```

```python
def count_clicks(impression_string):
    if pd.isna(impression_string):
        return 0, 0

    clicked = 0
    non_clicked = 0

    for token in impression_string.split():
        if token.endswith("-1"):
            clicked += 1
        elif token.endswith("-0"):
            non_clicked += 1

    return clicked, non_clicked


train_click_counts = train_behaviors["impressions"].apply(count_clicks)
dev_click_counts = dev_behaviors["impressions"].apply(count_clicks)

train_behaviors["num_clicked"] = train_click_counts.apply(lambda x: x[0])
train_behaviors["num_non_clicked"] = train_click_counts.apply(lambda x: x[1])

dev_behaviors["num_clicked"] = dev_click_counts.apply(lambda x: x[0])
dev_behaviors["num_non_clicked"] = dev_click_counts.apply(lambda x: x[1])

print("Train clicked items:", train_behaviors["num_clicked"].sum())
print("Train non-clicked exposed items:", train_behaviors["num_non_clicked"].sum())

print("Dev clicked items:", dev_behaviors["num_clicked"].sum())
print("Dev non-clicked exposed items:", dev_behaviors["num_non_clicked"].sum())
```
